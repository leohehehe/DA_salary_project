```python
import pandas as pd
pd.options.display.max_columns = None
pd.options.display.max_rows = None
```

```python
df = pd.read_csv('DA_job_dataset.csv',index_col = [0])
df = df[df['Salary Estimate'] != '-1']
df.head(3)
```

# Salary Parsing & Get dependent Variable Avg_salary

```python
df['hourly'] = df['Salary Estimate'].apply(lambda x: 1 if 'per hour' in x.lower() else 0)

Salary = df['Salary Estimate'].apply(lambda x: x.split('(')[0])
```

```python
#Change the Salary format from $20k-30k to two different columns: 20000 & 30000
Salary_num = Salary.apply(lambda x: x.replace('$','').replace('K','000').replace('Per Hour',''))
df['min_salary'] = Salary_num.apply(lambda x: int(x.split('-')[0]))
df['max_salary'] = Salary_num.apply(lambda x: int(x.split('-')[1]))
```

```python
#change hourly Salary to annual Salary
df['min_salary'] = df.apply(lambda x: x.min_salary*2000 if x.hourly==1 else x.min_salary,axis=1)
df['max_salary'] = df.apply(lambda x: x.max_salary*2000 if x.hourly==1 else x.max_salary,axis=1)
df[df['hourly']==1][['min_salary','max_salary']]
```

```python
#Calculate the Avg_salary
df['avg_salary'] = (df['min_salary']+df['max_salary'])/2
```

# Company Name text only

```python
df['Company_name_text'] = df.apply(lambda x: x['Company Name'][:-4] if int(x['Rating']) > 0 else x['Company Name'],axis= 1)
```

# State field

```python
df['State'] = df['Location'].apply(lambda x: x[-2:])
```

```python
df['State'].value_counts()
```

```python
#change the United States row to unknown 
df['State'].loc['es'] = 'Unknown'
```

# Age of company

```python
df['Age of company'] = df['Founded'].apply(lambda x: 2020-x if x != -1 else -1)
```

```python
df['Age of company'].value_counts()
```

# Job Description Parsing

```python
#python
df['python_y'] = df['Job Description'].apply(lambda x: 1 if 'python' in x.lower() else 0)

#SQL
df['sql_y'] = df['Job Description'].apply(lambda x: 1 if 'sql' in x.lower() else 0)

#AWS
df['aws_y'] = df['Job Description'].apply(lambda x: 1 if 'aws' in x.lower() else 0)

#Spark
df['spark_y'] = df['Job Description'].apply(lambda x: 1 if 'spark' in x.lower() else 0)

#Tableau
df['tableau_y'] = df['Job Description'].apply(lambda x: 1 if 'tableau' in x.lower() else 0)

#excel
df['excel_y'] = df['Job Description'].apply(lambda x: 1 if 'excel' in x.lower() else 0)

#R Studio
df['r_studio_y'] = df['Job Description'].apply(lambda x: 1 if ' R ' in x else 0)

#Power BI
df['power_bi_y'] = df['Job Description'].apply(lambda x: 1 if 'power bi' in x.lower() else 0)
```

```python
df.head()
```

# Job Description Length

```python
df['desc_len'] = df['Job Description'].apply(lambda x: len(x))
```

# Senority Column

```python
def seniority(title):
    if 'sr' in title.lower() or 'senior' in title.lower() or 'lead' in title.lower() or 'principal' in title.lower():
        return 'senior'
    elif 'jr' in title.lower() or 'junior' in title.lower():
        return 'junior'
    else:
        return 'na'
    
df['seniority'] = df['Job Title'].apply(seniority)
```

# To csv file

```python
df.to_csv('DA_job_dataset_cleaned.csv')
```
